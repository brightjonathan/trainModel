Load the sequence

```python
#this code combines text files from multiple directories into a single text file
with open("../output/combined_text.txt", "r", encoding="utf-8") as f:
    text_sequence = f.read()

len(text_sequence) # Output the length of the combined text sequence
```

```python
#this code allows you to run the script from the parent directory
import sys
sys.path.append('..')
```

```python
#with the help of the minbpe library, we can tokenize the text sequence and sys we imported
from minbpe.basic import BasicTokenizer

tokenizer = BasicTokenizer()
tokenizer.train(text_sequence, vocab_size=1024)
```

```python
#this code vocabulary is a dictionary mapping tokens to their indices
# and can be used to convert text into a sequence of token indices
vocab = tokenizer.vocab
vocab
```

```python
###TESTING
#encoding a text sequence into token indices
tokenizer.encode("Beginning of the text sequence")
```

```python
###TESTING
#decoding a sequence of token indices back into text
tokenizer.decode([66, 101, 103, 272, 110, 351, 662, 411, 495, 120, 275, 384, 113, 117, 686, 101])
```

```python
#This code will help you to add special tokens to the tokenizer
#The max_vocab_id is the last token ID in the vocabulary

max_vocab_id = list(tokenizer.vocab.keys())[-1]
tokenizer.special_tokens = {
    "<|startoftext|>": max_vocab_id + 1,
    "<|separator|>": max_vocab_id + 2,
    "<|endoftext|>": max_vocab_id + 3,
    "<|unk|>": max_vocab_id + 4,
    "<|padding|>": max_vocab_id + 5,
}
```

```python
#got 669k plus tokens. most open source models have trillions of tokens because they are trained on billions of text sequences
#this code will help you to encode the entire text sequence into token indices
len(tokenizer.encode(text_sequence))  #669642
```

```python
#saving it in the output directoty so that it can be use later
tokenizer.save(file_prefix="../output/tokenizer/my_tokenizer")
```
